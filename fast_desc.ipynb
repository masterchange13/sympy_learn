使用sympy实现基本的最速下降法，作为演示使用

```python
import sympy as sp 

# 定义参数
x1, x2 = sp.symbols('x1 x2')

# 定义目标函数
f = x1**2 + x2**3 + 2*x1*x2
```

```python
# 定义参数
lr = 1e-3
espi = 1e-6
x0 = sp.Matrix([100, 100])
```

```python
# 定义梯度函数,输入函数和变量，获取当前点梯度向量
def grad_f(f, x):
    length = len(f.free_symbols)
    #  定义梯度向量
    grip = sp.Matrix([0]*length)
    
    for i, symbol in enumerate(f.free_symbols):
        grip[i] = sp.diff(f, symbol)
    
    geard = grip.subs([(symbol, value) for symbol, value in zip(f.free_symbols, x)])
    return geard
```

```python
# 定义最速下降法函数
def steepest_descent(f, x0, lr, espi):

    dk = grad_f(f, x0)    

    while(dk.norm() > espi):
        # 计算梯度
        dk = grad_f(f, x0)
        # 更新参数
        x0 = x0 - lr * dk

    return x0, f.subs([(symbol, value) for symbol, value in zip(f.free_symbols, x0)])
```

```python
# start
steepest_descent(f, x0, lr, espi)
```
